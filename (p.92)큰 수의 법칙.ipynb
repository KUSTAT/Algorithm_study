```python
# N, M, K를 입력 받기
N, M, K = map(int, input().split())

# 배열을 리스트로 입력 받기
array = list(map(int, input().split()))

# 리스트에서 가장 큰 수 찾기
first = max(array)

# 리스트에서 가장 큰 숫자 제거(큰 숫자의 동점자가 있을 경우 원소가 하나만 빠지므로 동점자는 남아있음)
array.remove(first)

# 원소가 하나 빠진 리스트에서 가장 큰 숫자 찾기
second = max(array)

# (K+1)개 단위로 숫자가 반복되므로 (K+1)의 몫과 나머지를 통해 답 구하기
answer = M//(K+1)*(first*3+second) + M%(K+1)*first

print(answer)
```
