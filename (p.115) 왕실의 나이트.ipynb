```python
initial = input()

# x 좌표값(알파벳 a:1,b:2, ... h:8로 변환 후 할당)
x = ord(initial[0])-96

# y 좌표값
y = int(initial[1])

# 도착 가능 지점의 개수를 저장하는 변수 선언
count = 0

# 나이트가 움직일 수 있는 모든 경우의 수에 대한 리스트를 만들고 for문을 돌림
for move_knight in [[2,1],[1,2],[2,-1],[-1,2],[-2,1],[1,-2],[-2,-1],[-1,-2]]:
  
  # X의 이동 후 도착지점 구하기
  x_final = x + move_knight[0]

  # y의 이동 후 도착지점 구하기
  y_final = y + move_knight[1]

  # x,y의 이동후 도착지점이 8*8 체스판 안에 들어가는 지 if문을 통해 체크
  if x_final in [1,2,3,4,5,6,7,8] and y_final in [1,2,3,4,5,6,7,8] :
    
    # 도착지점이 8*8 체스판 안에 잘 들어가면 count에 1 더하기
    count += 1

# 정답 출력
print(count)
```
