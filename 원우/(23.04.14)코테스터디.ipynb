## p.321 럭키스트레이트, 구현문제

```python
# input 받기, 문자열로 받음, 슬라이싱하기 위해서
N = input()

# 정수의 자릿수 구하기
len_ = len(N)

# 왼쪽 자리수 합 저장
left_ = 0

# 오른쪽 자리수 합 저장
right_=0

# 왼쪽 부분의 자리 수 합 구하기
for i in range(len_ // 2):
    left_ += int(N[i])
    
# 오른쪽 부분의 자리 수 합 구하기
for i in range(len_//2):
    right_ += int(N[len_-i-1])
    
# 왼쪽 오른쪽 합이 서로 같으면 럭키 스트레이트!
if left_ == right_:
    print("LUCKY")
else :
    print("READY")
```

## p.322, 문자열 재정렬, 구현문제

### 마지막에 리스트를 문자열로 변환하여 출력하는 것 잘 못했음 익혀놓기
### ' '.join(answer) 로 리스트를 문자열로 변환!

```python
# 문자열 입력받기
S = input()

# 결과를 입력받을 빈 리스트 만들기
answer = []

# 숫자 합을 계산할 변수 만들기
num_sum = 0

# 문자를 하나씩 확인!
for x in S:
    if x.isalpha():
        answer.append(x)
    else :
        num_sum += int(x)

# answer에 들어간 알파벳을 오름차순으로 정렬
answer.sort()

# 숫자가 하나라도 존재하는 경우 가장 뒤에 삽입
if num_sum != 0:
    answer.append(str(num_sum))

# 최종결과 리스트 출력
print(''.join(answer))
```

## p.339 Q15) 특정거리의 도시찾기, DFS/BFS

```python
from collections import deque
from sys import stdin
#input = stdin.readline

# 도시의 개수, 도로의 개수, 거리 정보, 출발 도시 번호
N, M, K, X = map(int, input().split())

# 각 노드별 연결된 노드를 저장할 리스트 만들기
graph = [[] for _ in range(N+1)]

# 모든 도로 정보를 graph에 입력받기
for _ in range(M):
    A,B = map(int, input().split())
    graph[A].append(B)

# 모든 도시에 대한 최단 거리 -1로 설정해두기
short_d = [-1]*(N+1)

# 출발 도시까지의 거리는 0으로 설정
short_d[X] = 0

# BFS(너비우선탐색) 수행

# 큐 만들기
q = deque([X])

while q:
    # 현재 도시 큐에서 뽑기
    now = q.popleft()
    
    # 현재 도시에서 이동할 수 있는 모든 도시를 확인
    for next_node in graph[now]:
        # 아직 방문하지 않은 도시라면
        if short_d[next_node] == -1:
            # 최단 거리 갱신
            short_d[next_node] = short_d[now] + 1
            q.append(next_node)

# 최단 거리가 K인 모든 도시의 번호를 오름차순으로 출력
# check가 True면 최단거리가 K인 도시가 존재하는 것, check가 False면 최단거리가 K인 도시가 없는 것
check = False
# short_d에 K까지의 거리가 들었는지 탐색, 오름차순으로 for문 돌렸!
for i in range(1, N+1):
    if short_d[i] == K:
        # 거리가 K인거 찾으면 노드 숫자 내보내기
        print(i)
        # if문 안으로 한 번이라도 들어오면 check
        check = True
        
if check == False:
    print(-1)
```

## p.341 연구소, DFS/BFS

### 구현 잘 안되서 고생... 어렵...
### 재귀함수 쓰는거 왤케 어렵냐고 제에에에에에엔자아아아아앙
### 잘 모르겠어서 책이랑 똑같이 했는데 시간초과뜸...

```python
from sys import stdin
# 입력 빠르게 받아야함
input = stdin.readline

# 맵 크기 input() 받기
n, m = map(int, input().split())
data = [] # 초기 맵 리스트
temp = [[0]*m for _ in range(n)] # 벽을 설치한 뒤의 맵리스트

for _ in range(n):
    data.append(list(map(int, input().split())))
    
# 4가지 이동 방향에 대한 리스트
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

# 결과값 저장
result = 0

# 깊이 우선 탐색(DFS)를 이용해서 각 바이러스가 사방으로 퍼지게 하기
def virus(x,y):
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        # 바이러스가 칸 밖으로는 나갈 수 없기 때문에 칸 안에서만 퍼진다고 if문 걸어줌
        if nx >= 0 and nx < n and ny >= 0 and ny < m:
            # 옮겨진 곳에 바이러스가 없다면!
            if temp[nx][ny] == 0:
                # 해당 위치에 바이러스를 배치하고 다시 재귀적으로 수행
                temp[nx][ny] = 2
                # 재귀함수~~~~
                virus(nx, ny)

# 현재 맵에서 안전 영역의 크기를 계산하는 함수
def get_score():
    # score 변수를 정의하기
    score = 0
    # 이중 for문을 돌면서 안전영역(0인 지점)의 개수를 계산
    for i in range(n):
        for j in range(m):
            if temp[i][j] == 0:
                score += 1
    return score
                
# 깊이 우선 탐색(DFS)를 이용해 울타리를 설치하면서 매번 안전 영역의 크기를 계산(완전 탐색!!!)
def dfs(count):
    global result
    # 울타리가 3개 설치된 경우
    if count == 3:
        for i in range(n):
            for j in range(m):
                temp[i][j] = data[i][j]
        # 각 바이러스의 위치에서 전파 진행
        for i in range(n):
            for j in range(m):
                if temp[i][j] == 2:
                    virus(i,j)
        # 안전 영역의 최댓값 계산
        result = max(result, get_score())
    # 빈 공간에 울타리 설치했다가 해체했다가 반복
    # 이중 for문을 돌면서 전체 탐색
    for i in range(n):
        for j in range(m):
            # 빈 칸이면
            if data[i][j] == 0:
                # 울타리 설치
                data[i][j] = 1
                # 울타리 개수 올리기
                count += 1
                # 재귀함수 실행
                dfs(count)
                # 울타리 해체
                data[i][j] = 0
                # 울타리 개수 줄이기
                count -= 1
dfs(0)
print(result)
```

## p.367 Q27) 정렬된 배열에서 특정 수의 개수 구하기

### 이진 탐색을 두 번 써서 풀어야 했었음...
### 책에서와는 다르게 재귀함수를 사용하지 않고 짰음
### 파이썬 이진 탐색 라이브러리 bisect도 알아놓기

```python
N, x = map(int, input().split())
array = list(map(int, input().split()))

# 처음 등장한 인덱스를 찾는 이진 탐색 알고리즘 짜기
def first(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        
        # 해당 값을 가지는 원소 중에서 가장 왼쪽에 있는 경우에만 인덱스 반환
        # and 왼쪽 : mid 인덱스보다 한 칸 왼쪽으로 움직였을 때 target보다 작아져야함
        # and 오른쪽 : mid 인덱스는 딱 target이어야 함
        if (mid == 0 or target > array[mid-1]) and array[mid] == target:
            return mid
        
        # 중간점의 값보다 찾고자 하는 값 같거나 작은 경우 왼쪽 확인
        # (기초적인 이분탐색 알고리즘과는 다르게 같거나가 추가된 이유:위에서 같아도 제일 왼쪽에 있는 경우만 인덱스 반환했기 때문)
        elif array[mid] >= target:
            end = mid - 1
        
        # 중간점의 값보다 찾고자 하는 값이 큰 경우 오른족 확인
        else:
            start = mid + 1
    return None

# 마지막 등장한 인덱스를 찾는 이진 탐색 알고리즘 짜기
def last(array, target, start, end):
    while start <= end:
        
        mid = (start + end) // 2
        
        # 해당 값을 가지는 원소 중에서 가장 오른쪽에 있는 경우에만 인덱스 반환
        # and 왼쪽 : mid 인덱스보다 한 칸 오른쪽으로 움직였을 때 target보다 커저야 함
        # and 오른쪽 : mid 인덱스는 딱 target이어야 함
        if (mid == N-1 or target < array[mid+1]) and array[mid] == target:
            return mid
        
        # 중간점의 값보다 찾고자 하는 값 같거나 작은 경우 왼쪽 확인
        # 헷갈렸었었음!!!(first와는 다르게 등호가 빠지는 이유 같은값이면 오른쪽으로 전진해야하기 때문)
        elif array[mid] > target:
            end = mid - 1
        
        # 중간점의 값보다 찾고자 하는 값이 큰 경우 오른족 확인
        else:
            start = mid + 1
    return None
             
# 정렬된 수열에서 값이 x인 원소의 개수를 세는 함수
# 이미 정렬된 수열이기 때문에 이진탐색으로 시간 복잡도를 줄일 수 있음
def count_by_value(array, x):
    # 데이터의 개수
    n = len(array)
        
    # 인덱스 0~(n-1) 사이에 x가 처음 등장한 인덱스 계산
    a = first(array, x, 0 ,n-1)
        
    # 수열에 x가 존재하지 않는 경우
    if a == None:
        return 0 # 개수가 0개이므로 0 반환
            
    # 인덱스 0~(n-1) 사이에 x가 처음 등장한 인덱스 계산
    b = last(array, x, 0, n-1)
    
    # a,b를 활용하여 x의 총 개수를 반환
    return b-a+1
            
# 값이 x인 데이터의 개수 계산 
count = count_by_value(array, x)

# 값이 x인 원소가 존재하지 않는다면
if count == 0:
    print(-1)
# 값이 x인 원소가 존재한다면
else :
    print(count)
```

## p.368 Q28) 고정점 찾기

```python
N = int(input())
array = list(map(int, input().split()))

# 책에서 사용한 이진탐색 알고리즘
def search_fixed_point(array, start, end):
    while start <= end:
        mid = (start + end) // 2
        # 고정점을 찾은 경우 중간점 인덱스 반환
        if array[mid] == mid:
            return mid
        # 고정점이 안됐는데 인덱스 값보다 실제값이 더 클 경우 왼쪽 탐색해야 함
        elif array[mid] > mid:
            end = mid - 1
        # 고정점이 안됐는데 인덱스 값보다 실제값이 더 작을 경우 오른쪽 탐색해야 함
        else:
            start = mid + 1
    return None

fixed_point = search_fixed_point(array, 0, N-1)
if fixed_point == None :
    print(-1)
else :
    print(fixed_point)
```
