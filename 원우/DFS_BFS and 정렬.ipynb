## p. 149, 음료수 얼려먹기


```python
N, M = map(int, input().split())
ice_frame = [list(map(int, input())) for _ in range(N)]
```

```python
# 방문 기록용 2중 리스트 만들기
visited = [[0]*M for _ in range(N)]

print(visited)
print(ice_frame)
```

```python
def dfs(ice_frame, x, y, visited):
  print(visited)
  # 현재 노드를 방문 처리(1로 방문처리)
  visited[x][y] = 1

  # 이동할 네 방향 정의
  dx = [0, -1, 0, 1]
  dy = [-1, 0, 1, 0]

  # 현재 노드와 연결된 다른 노드를 재귀적으로 방문
  for i in range(4) :
    
    # 방문할 지점을 변수로 설정
    nx = x+dx[i]
    ny = y+dy[i]

    print("({},{})".format(nx,ny))
    
    # x_new와 y_new가 N*M 범위 안에 들어 올 때만 탐색
    if nx in range(N) and ny in range(M):
      
      # 얼음틀에서 0인 지점이고 방문하지 않은 지점일 때 재귀함수 실행
      if ice_frame[nx][ny] == 0 and visited[nx][ny] != 1 :
        visited[nx][ny]=1
        dfs(ice_frame, nx, ny, visited)
    
    else :
      continue
```

```python
count = 0

for i in range(N):
  for j in range(M) :
    if ice_frame[i][j] == 0 and visited[i][j] == 0 :
      print("count : {}".format(count))
      dfs(ice_frame, i, j, visited)
      count += 1
```

### p.152, 미로탈출

```python
N, M = map(int, input().split())
maze = [list(map(int, input())) for _ in range(N)]
```

```python
maze_visit = [[0]*M for _ in range(N)]
```

```python
from collections import deque

def bfs(maze, x, y, maze_visit):
  
  # 거리 기록용 변수
  distance = 0

  # queue 구현을 위해 deque 라이브러리 사용
  queue = deque()
  queue.append((x,y))
  
  # 현재 노드를 방문처리
  maze_visit[x][y] = 1
  
  # 이동할 네 방향 정의
  dx = [0, -1, 0, 1]
  dy = [-1, 0, 1, 0]

  # 큐가 빌 때까지 반복
  while queue :

    # 큐에서 하나의 원소를 뽑아 출력
    x, y = queue.popleft()

    # 큐에서 뽑은 원소에서 갈 수 있는 네 방향 탐색
    for i in range(4) :
      nx = x + dx[i]
      ny = y + dy[i]
      
      # 진행 방향이 더 나아갈 수 없는 곳 못감
      if nx not in range(N) or ny not in range(M) :
        continue
      
      # 괴물이 있는 공간 못감
      if maze[nx][ny] == 0 :
        continue

      # 방문 안했던 공간만 최단 거리에 기록
      if maze_visit[nx][ny] == 0 :
        maze_visit[nx][ny] = maze_visit[x][y] + 1
        queue.append((nx,ny))

  return maze_visit[N-1][M-1]

# BFS를 수행한 결과 출력
print(bfs(maze, 0, 0, maze_visit))
```

## p.178 위에서 아래로

```python
# 책의 퀵정렬 소스코드를 내림차순으로 변형

# 문제 조건 인풋 받기
M = int(input())
seq_my = [int(input()) for i in range(M)]

# 퀵소트 함수
def quick_sort(array, start, end):
  if start >= end: #원소가 한개인 경우 종료
    return
  pivot = end #마지막 원소를 피벗으로 두기
  left = start
  right = end-1

  while left <= right:
    # 왼쪽부터는 피벗보다 작은 데이터를 찾을 때 까지 반복
    while left < end and array[left] >= array[pivot] :
      left += 1
    # 오른쪽부터는 피벗보다 큰 데이터를 찾을 때 까지 반복
    while right >= start and array[right] <= array[pivot] :
      right -= 1
    if left > right: # left, right 엇갈릴 때 left와 피벗 교체
      array[left], array[pivot] = array[pivot], array[left]
    else: # 엇갈리지 않았다면 큰 데이터와 작은 데이터를 교체
      array[left], array[right] = array[right], array[left]

  # 분할 이후 왼쪽 부분과 오른쪽 부분에서 각각 정렬수행
  # left와 피벗 교체 했기 때문에 책과는 달라짐
  quick_sort(array, start, left-1)
  quick_sort(array, left+1, end)


# 퀵소트 함수 실행
quick_sort(seq_my, 0, len(seq_my)-1)

# 정답 출력
print(*seq_my, sep=" ")
```

## p.182 두 배열의 원소 교체

```python
N, K = map(int, input().split())
arrA = list(map(int, input().split()))
arrB = list(map(int, input().split()))

# arrA 오름차순 정렬
arrA = sorted(arrA)

# arrB는 내림차순 정렬
arrB = sorted(arrB, reverse=True)

# K번 탐색
for i in range(K):
  # arrA의 i번째 원소가 arrB의 i번째 원소보다 작을 때는 arrA가 커지기 위해 교체 필요 있음
  if arrA[i] < arrB[i] :
    arrA[i], arrB[i] = arrB[i], arrA[i]
  # arrA의 i번째 원소가 arrB의 i번째 원소보다 크거나 같을 때는 교체 필요도 없고 for문에서 더 탐색할 필요도 없으므로 break
  else :
    break

print(sum(arrA))
```
