# 랜선자르기 백준 164

```python
K, N = map(int, input().split())
lan_cables = [int(input()) for _ in range(K)]

# 시작값 1, 끝값 주어진 랜선 길이 중 가장 긴 값으로 잡음
start = 1
end = max(lan_cables)

# 시작값과 끝값이 같을 때까지 반복
while start <= end:
    # 중간값 탐색
    mid = (start + end) // 2

    # 각 케이블에서 mid를 나눈 몫을 다 더해줌 => 만들어지는 랜선 개수가 나옴
    total_cables_num = sum([cable // mid for cable in lan_cables])

    # N보다 크거나 같을 때 오른쪽 탐색
    if total_cables_num >= N:
        start = mid + 1
    # 반대일 때 왼쪽 탐색
    else:
        end = mid - 1

print(end)
```
