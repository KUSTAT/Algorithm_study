```python
# N, K 입력 받기
N, K = map(int, input().split())

# 최소 횟수 기록하기 위한 변수 만들기
count = 0

# N이 1이 아니면 while문을 계속 돌림
while N != 1 :

  # N이 K로 나눠지면 나누기를 수행하여 N에 저장
  if N%K == 0 :
    N = int(N/K)
  
  # N이 K로 나눠지지 않으면 N-1을 N에 저장
  else :
    N -= 1
  
  # while문이 한 바퀴 돌 때 마다 수행 횟수 기록
  count += 1

print(count)
```
