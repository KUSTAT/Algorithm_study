```python
# N, M 입력 받기
N, M = map(int, input().split())

# 각 행마다 가장 작은 수를 입력받기 위한 리스트를 만듬(리스트 길이는 열개수 N)
min_row = [0 for i in range(N)]

# 반복문을 N회 반복하여 각 행마다 가장 작은 숫자 찾기
for i in range(N) :

  # 행을 리스트로 입력 받기
  new = list(map(int, input().split()))
  
  # 행에서 가장 작은 숫자 찾기
  min_row[i] = min(new)

# 각 행마다 가장 작은 수를 입력받은 리스트에서 가장 큰 숫자를 출력하여 정답 도출  
print(max(min_row))
```
