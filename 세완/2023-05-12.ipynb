```python
### Q6. 무지의 먹방 라이브 (시간 초과)
def solution(food_times, k):
    answer = 0
    l = len(food_times)
    temp = True
    if sum(food_times) <= k:
        answer = -1
    else:
        while temp:
            for i in range(l):
                if food_times[i] == 0:
                    pass
                elif k != 0:
                    food_times[i] -= 1
                    k -= 1
                else:
                    answer = i+1
                    temp = False
                    break
    return answer
```

```python
### Q6. 무지의 먹방 라이브 (몇개 시간초과 : 답지랑 아이디어는 비슷한거같은데 왜?)
import heapq
def solution(food_times, k):
    answer = 0
    out = True
    l = len(food_times)
    q = []
    i = 0
    temp = 0
    # (시간, 음식 인덱스)로 구성된 PQ
    for x in food_times:
        i += 1
        heapq.heappush(q, (x, i))
    # k초안에 모든 음식을 다 먹을수 있는 케이스
    if sum(food_times) <= k:
        answer = -1
        return answer
    while out:
        m = min(q)[0]
        if temp == m:
            heapq.heappop(q)
            continue
        div = k//len(q)
        if div >= (m-temp):
            k -= (m-temp)*len(q)
            temp = m
        else:
            k = k % len(q)
            arr = sorted(q, key = lambda x : x[1])
            answer = arr[k][1]
            out = False
    return answer
```

```python
#### Q11. 뱀
# 데이터 입력
n = int(input())
k = int(input())
graph = [[0]*n for _ in range(n)]
snake = [[0]*n for _ in range(n)]
for _ in range(k):
    r, c = map(int, input().rstrip().split())
    graph[r-1][c-1] = 1
l = int(input())
move = []
for _ in range(l):
    x, c = input().rstrip().split()
    x = int(x)
    move.append((x, c))
# 구현
from collections import deque
def dummy(snake, graph, n, k, l):
    snake[0][0] = 1
    path = deque()
    head = [0, 0]
    path.append([0, 0])
    # 시간 초
    answer = 0
    time = 0
    end = False
    # 방향
    dr = [0, 1, 0, -1]
    dc = [1, 0, -1, 0]
    idx = 0
    d = (dr[idx], dc[idx])
    for x in move:
        new_time, direction = x[0], x[1]
        t = new_time - time
        time = new_time
        for j in range(t):
            # 머리 이동
            head[0] += d[0]
            head[1] += d[1]
            # 머리가 범위 안에 있고 몸통이랑 안 부딪혔을 경우 >> 머리 이동
            if 0 <= head[0] < n and 0 <= head[1] < n and snake[head[0]][head[1]] == 0:
                path.append([head[0], head[1]])
                snake[head[0]][head[1]] = 1
                # case1 : 사과가 있는 경우 >> 해당 사과 없애고 꼬리 이동 X
                if graph[head[0]][head[1]] == 1:
                    graph[head[0]][head[1]] = 0
                    answer += 1
            
                # case2 : 사과가 없는 경우 >> 꼬리 이동
                else:
                    r, c = path.popleft()
                    snake[r][c] = 0
                    answer += 1
            else:
                return answer+1
        # 방향 전환
        if direction == 'D':
            idx = (idx+1) % 4
            d = (dr[idx], dc[idx])
        else:
            idx = (idx-1) % 4
            d = (dr[idx], dc[idx])
    while True:
        # 머리 이동
        head[0] += d[0]
        head[1] += d[1]
        # 머리가 범위 안에 있고 몸통이랑 안 부딪혔을 경우 >> 머리 이동
        if -1 < head[0] < n and -1 < head[1] < n and snake[head[0]][head[1]] == 0:
            path.append([head[0], head[1]])
            # case1 : 사과가 있는 경우 >> 해당 사과 없애고 꼬리 이동 X
            if graph[head[0]][head[1]] == 1:
                graph[head[0]][head[1]] = 0
                answer += 1
            
            # case2 : 사과가 없는 경우 >> 꼬리 이동
            else:
                r, c = path.popleft()
                snake[r][c] = 0
                answer += 1
        else:
            answer += 1
            return answer 

ans = dummy(snake, graph, n, k, l)
print(ans)
```

```python
# Q18. 괄호 변환
# 올바른 괄호 문자열인지 판별
def right(p):
    cnt = 0
    for i in range(len(p)):
        if p[i] == '(':
            cnt += 1
        else:
            cnt -= 1
        if cnt < 0:
            return False
    return True
# 균형잡힌 괄호 문자열인지 판별
def balance(p):
    cnt = 0
    for i in range(len(p)):
        if p[i] == '(':
            cnt += 1
        else:
            cnt -= 1
    if cnt == 0:
        return True
    else:
        return False
# 괄호 뒤집기
def reverse(p):
    answer = ''
    for i in range(len(p)):
        if p[i] == '(':
            answer += ')'
        else:
            answer += '('
    return answer
# 정답 도출
def solution(p):
    answer = ''
    cnt = 0
    # 이미 올바른 괄호 문자열인 경우
    if right(p):
        return p
    # 아닌경우
    # 2번 수행
    for i in range(1, len(p)):
        temp = p[:(i+1)]
        if balance(temp):
            u = temp
            v = p[(i+1):]
            break
    # 3번 수행
    if right(u):
        answer = u + solution(v)
    # 4번 수행
    else:
        answer = '(' + solution(v) + ')' + reverse(u[1:-1])
    return answer
```
