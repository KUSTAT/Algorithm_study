```python
# "북, 서, 남, 동" 반시계 방향으로 바라보는 곳의 위치 리스트 만들기
see_list=[[0,1],[-1,0],[0,-1],[1,0]]

count = 0
decision = 0

# 움직이는 함수
def move_game(N, M, A, B, see, map_info, count, decision):
  
  # 원래 카운트를 입력 받아놓기
  origin_count = count

  for i in range(4) :
    
    # 지정한 범위 안에서만 움직이게 if문을 걸어줌
    if (A + see_list[(see+i//4)][0]) in range(N) or (B + see_list[(see+i)//4][1]) in range(M):
      
      # 육지인 경우 이동하도록 if문을 걸어줌
      if map_info[A + see_list[(see+i)//4][0]][B + see_list[(see+i)//4][1]] == 0:
        
        # 이동하기전에 땅을 2로 만들어서 표시
        map_info[A][B] = 2
        
        # A,B 를 실제로 이동
        A += see_list[(see+i)//4][0]
        B += see_list[(see+i)//4][1]
        
        # 카운트 올려주기
        count += 1
        
        break
      break
    break
  
  # count가 올라가지 않은 경우
  if origin_count == count :
    
    # decision=1로 바꿔주기
    decision = 1
  return map_info, count, A, B, decision

while decision != 1 :
  map_info, count, A, B, decision = move_game(N, M, A, B, see, map_info, count, decision)
  print(decision)
  print(map_info)
  print(count)

print(map_info)
print(count)    
#       while map_info[A + see_list[(see+2)//4]][B + see_list[(see+2)//4]] != 0 :
#         map_info[A + see_list[(see+2)//4]][B + see_list[(see+2)//4]] == 0




# map_info[A] map_info[B]
```
